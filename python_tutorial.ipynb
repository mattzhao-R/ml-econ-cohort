### What is a function?
In python, you can call functions that will do things. That's very vague so here's an example:

### What is a variable? 
We store information in python in variables. Here's an example. Please name your variables something relevant to what they are.
lst contains a list, which is a type of data structure in python. lists contain multiple things. 

```
number = 3
lst = np.random.random(6) # generates 50 random numbers between 0 and 1 (see uniform distribution)
print(number)
print(lst)
```

You can perform functions on variables, and you can assign the output to the same or another variable or you can leave it unsaved. What's great about python is this property called vectorization - you can call np.log on a list or multiply it by 3 and it will give you the natural log of every element in a list. You cannot do this in every programming language. 

```
print(np.sqrt(number))
log_lst = np.log(lst)
print(log_lst)
times_3_lst = lst * 3
print(times_3_lst)
```

A for loop allows you to iterate through the contents of a list. See the two examples of how to do so with a for loop that produce the same output. We can either reference the objects in a list, or we can reference the position an object is in a list. 

```
animals = ['dog', 'cat', 'bird']
for animal in animals:
    print("The plural of " + animal + " is " + animal + "s")
```

```
animals = ['dog', 'cat', 'bird']
for i in range(len(animals)):
    print("The plural of " + animals[i] + " is " + animals[i] + "s")
```
